```python
# First, let's load up our payphone data.
# I collected the nearest 1000 payphones to Adelaide from http://envinsaonline.mapinfo.com.au/ppol/.

import json
that = open('adelaide1k.json','r')
feats = json.loads(that.read())[u'features']
```

```python
# Now let's put it into a dataframe.
import pandas as pd
import numpy as np

# Here the longlist represents a list of rows, while short is a single row.
longlist = []
cats = [u'SECTOR', u'SSC_NAME', u'FEDERAL_ELECTORATE', u'MI_PRINX', u'NEXT_DISTANCE', u'STATE_ELECTORATE', u'distanceToFeature', u'STYLE', u'LONGITUDE', u'PHONE_ATTRIBUTES', u'STATE', u'POSTCODE', u'ADDRESS', u'LATITUDE', u'CABINET_ID', u'LGA_NAME', u'NEXT_PAYPHONE']
for f in feats:
    short = []
    for c in cats:
        try:
            short.append(f[u'properties'][c])
        except:
            short.append(np.nan)
    longlist.append(short)
df = pd.DataFrame(longlist)
df.columns = cats
df.head()
```

```python
# First, a quick check to see if the locations make sense.
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style('white')
plt.scatter(df['LONGITUDE'],df['LATITUDE'],alpha=0.05)
```

```python
# If anyone knows Adelaide's geography they will know that this looks good!
# Now the data comes ready-made with a few classifications.
# First, let's investigate any electorate bias.
df['FEDERAL_ELECTORATE'].value_counts().plot(kind='bar')
```
