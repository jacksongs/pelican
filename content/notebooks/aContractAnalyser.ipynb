```python
# The imports...

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# Let's get our contract data in.

df = pd.read_csv('data.csv',index_col=0)
df = df.sort_index()
```

```python
# First let's have a look at how empty it is.
name = df[df['Name'] == 'Error Type:']
len(name)
```

```python
# Nice! Okay, so let's make a dataframe without these empty values.
# Note that I have checked that each of the rows I consider are 'empty' are actually empty by inspecting the name dataframe.

full = df[df['Name'] != 'Error Type:']

# And now let's make a dataframe just with columns we care about for now.

df1 = full[['Name','Title','Total inc GST','Name of Public Authority','Procurement Process','Completion Date','Starting Date','Execution Date']]
```

```python
# First, I'm interested in costs. Let's make another dataframe just with the contracts with a value.

dfc = df1[df1['Total inc GST'] != np.nan]

# And let's convert the cost to an integer.

dfc['Total inc GST'] = dfc['Total inc GST'].map(lambda x: int(str(x).translate(None, '$,')))

# Now let's see what a histogram of the costs looks like.
plt.hist(dfc['Total inc GST'].tolist(),50)
seaborn.set_style('dark')
```

```python
# Well that looks a bit crap. Let's try again, but only with values less than 2 million.

twomil = dfc[dfc['Total inc GST'] < 2000000]
plt.hist(twomil['Total inc GST'].tolist(),50)
seaborn.set_style('dark')
```

```python
# Clearly we see a cluster just over $500,000. Now let's try to do the same over each year.

# Finally, let's convert our data columns to actual dates.

twomil['Completion Date'] = pd.to_datetime(twomil['Completion Date'])
twomil['Starting Date'] = pd.to_datetime(twomil['Starting Date'])
twomil['Execution Date'] = pd.to_datetime(twomil['Execution Date'])
```

```python
# Now let's start plotting. First, grouping year in completion date.

comp = twomil[twomil['Completion Date'] > datetime.datetime(2005,1,1)]
comp = comp[twomil['Completion Date'] < datetime.datetime(2024,1,1)]
comp = comp.set_index('Completion Date')
compbyyear = comp['Total inc GST'].groupby(lambda x: x.year)

fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(16, 8), 
                         tight_layout=True)

for ax, grp in zip(axes.ravel(), compbyyear):
    ax.hist(grp[1],20, histtype='stepfilled')
    ax.set_yticks(np.arange(0, 200, 50))
    ax.set_ylim(0, 150)
    ax.set_title(grp[0])
    ax.axvline(500000, alpha=0.6, ls='--')
```

```python
# Now, let's try start date.

sta = twomil[twomil['Starting Date'] > datetime.datetime(2005,1,1)]
sta = sta[twomil['Starting Date'] < datetime.datetime(2015,1,1)]
sta = sta.set_index('Starting Date')
stabyyear = sta['Total inc GST'].groupby(lambda x: x.year)

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(16, 5), 
                         tight_layout=True)

for ax, grp in zip(axes.ravel(), stabyyear):
    ax.hist(grp[1],20, histtype='stepfilled')
    ax.set_yticks(np.arange(0, 200, 50))
    ax.set_ylim(0, 150)
    ax.set_title(grp[0])
    ax.axvline(500000, alpha=0.6, ls='--')
```

```python
# And now execution date.

exe = twomil[twomil['Execution Date'] > datetime.datetime(2005,1,1)]
exe = exe[twomil['Execution Date'] < datetime.datetime(2015,1,1)]
exe = exe.set_index('Execution Date')
exebyyear = exe['Total inc GST'].groupby(lambda x: x.year)

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(16, 5), 
                         tight_layout=True)

for ax, grp in zip(axes.ravel(), exebyyear):
    ax.hist(grp[1],20, histtype='stepfilled')
    ax.set_yticks(np.arange(0, 200, 50))
    ax.set_ylim(0, 150)
    ax.set_title(grp[0])
    ax.axvline(500000, alpha=0.6, ls='--')
```
