```python
# First, some imports.

from pattern.web import URL, Element, plaintext
import pandas as pd
import numpy as np
import json
```

```python
# Next, some preliminary steps.

# This is the base URL. We will be appending a number at the end of this.
base = 'https://www.tenders.sa.gov.au/tenders/contract/view.do?id='
```

```python
# This is the scraper

# This reads the backup data and puts its back into the 'row' dictionary
# This is the last number backed up.
### Change this if you are running from a backup file.
rerun = True

if rerun == False:
    # This is for the first time run.
    # We are going to be creating one big dictionary that will then be turned into a dataframe.
    row = {}
    n = 1
else:
    # This is for the backup run. 
    n = 1500
    backup = open('data%s.txt'%n,'r')
    row = json.load(backup)

# This will help tell us when to stop the scraper later.
stopcount = 0

# And now, the scraper!
while True:
    # This is the trigger to stop the scraper.
    if stopcount > 20:
        # There is a big chunk of contracts missing around the 1600s.
        # We keep the scraper running during this range, just in case they come back in future.
        if 1600 < n < 1800:
            pass
        else:
            print 'Ended at:',datetime.datetime.now(),', at id:',n
            break
    
    # This part backs up the data and gives us a status update every time the scraper gets to its 500th page.
    # The backup system is simply a text file with one big json object.
    if n % 500 == 0:
        print n,datetime.datetime.now()
        with open('data%s.txt'%str(n), 'w') as outfile:
            json.dump(row, outfile)
        outfile.close()
    
    # Here we create a new key in our dictionary for the current contract.
    row[n] = {}
    
    # This downloads the page and passes it to the pattern module.
    url = URL(string=base+str(n))
    html = url.download(timeout=10, cached=True, throttle=0, proxy=None, unicode=False)
    element = Element(html)
    
    # This is the auto stopper. If we get an Error 20 consecutive times, we'll end the scraper.
    try:
        if str(element.by_tag('h2')[0].content) == 'Error Type:':
            stopcount += 1
        else:
            stopcount = 0
    except:
        pass 
    
    # Here we grab the data.
    # First, the name. Note I have wrapped everything in try/except statements in case the scraper hits a page with a different format.
    try:
        row[n]['Name'] = element.by_tag('h2')[0].content
    except Exception as e:
        row[n]['Name'] = e # Here I record name errors in the Name column.
        
    # Second, each left hand value in the table - the category - and its corresponding right-hand value.
    # This required a bit of relative movement. This is perhaps this scraper's biggest weakness.
    # If the table format is different, the data won't be collected.
    try:
        for i in element.by_class('LIST_TITLE'):
            for u in i.children:
                if len(u) == 1:
                    row[n][str(plaintext(str(u))).strip()] = u.parent.parent.parent.by_tag('td')[1].content
    except Exception as e:
        row[n]['Error'] = e # Here I am recording errors in the Error column.
    
    # And of course, now increase n by one and start again...
    n += 1
    
# And now we put the whole freaking row in the dataframe!
data = pd.DataFrame.from_dict(row, orient='index') 
data.head()
```

```python
# And now we export the data as a csv...

data.to_csv('data.csv',encoding='utf-8')
```
